```python
# Global and local variable with differente name
x = "global"

def funct1():
  global x
  y = "local"
  x = x * 2
  print(x)
  print(y)

print("Global x =",x)
funct1()
print("Global x =",x)
```

```python
# Global and local variable with same name
a = 5

def funct2():
  a = 10
  print("local a =",a)

print("global a =",a)
funct2()
print("global a =",a)
```

```python
# Creating and using a Non-local variable

def outer():
  x = "local"

  def inner():
    nonlocal x
    x = "nonlocal"
    print("inner:",x)

  inner()
  print("outer:",x)

outer()
```
