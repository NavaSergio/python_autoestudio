```python
# Python Decimal
data1 = 0.1 + 0.2
print(data1)
data1 = 1.2 * 2.5
print(data1)
from decimal import Decimal as D
print(D('0.1') + D('0.2'))
print(D('1.2') * D('2.5'))
```

```python
# Python fractions
from fractions import Fraction as F
print(F(1.5))
print(F(5))
print(F(1,5))
```

```python
# Python math module
import math
print(math.pi)
print(math.cos(10))
print(math.log(10))
print(math.log10(10))
print(math.exp(10))
print(math.factorial(5))
print(math.sinh(10))
print(abs(-12.34))
```
