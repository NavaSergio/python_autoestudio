```python
# Creando una tupla vacia
tuple1 = ()
print(tuple1)

# creando una tupla con elementos enteros
tuple2 = (1, 2, 3)
print(tuple2)

# tupla con tipos de datos mesclados
tuple3 = (101, "Anirban", 20000.00, "HR Depot")
print(tuple3)
```

```python
# creacion de tuplas anidadas
tuple4 = ("points",[1, 4, 3], (7, 8 , 6))
print(tuple4)

# tuplas se pueden crear sin parentesis
# se llaman tuple packing (tuplas empaquetadas)
tuple5 = 101, "Anirban", 20000.00, "HR Depot"
print(tuple5)

# es posible desempacar una tubla
empid, empname, empsal, empdept = tuple5
print(empid)
print(empname)
print(empsal)
print(empdept)

print(type(tuple5))
```

```python
tuple1 = ('w','e','l','c','o','m','e')
print(tuple1)
print(tuple1[1])
print(tuple1[3])
print(tuple1[5])

#nested tuple
nest_tuple2 = ("point",[1,3,4],(8,7,9))
print(nest_tuple2)
print(nest_tuple2[0][3])
print(nest_tuple2[1][2])
print(nest_tuple2[2][2])
```

```python
# slicing tuple contents
tuple1 = ('w','e','l','c','o','m','e')
print(tuple1[1:3])
print(tuple1[:-3])
print(tuple1[3:])
print(tuple1[:])
```

```python
# los elementos de las tuplas son inmutables
tuple1 = ('w','e','l','c','o','m','e')
print(tuple1)
# tuple1[2]='x'  # error

# tuplas pueden ser reasignadas
tuple1 = ('g','o','o','d','b','y','e')
print(tuple1)
```

```python
# concatenacion de tuplas
tuple2 = ('w', 'e', 'l')
tuple3 = ('c', 'o', 'm', 'e')
print(tuple2)
print(tuple3)
print(tuple2 + tuple3)

print(("again",) *4)
```

```python
# deletion operation on a tuple
tuple4 = ('w','e','l','c','o','m','e')
print(tuple4)
# as inmutable so elements can not be deleted
# del tuple4[2]

# but can delete entire tuple
del tuple4
print(tuple4)
```

```python
# tuple methods
tuple5 = ('w','e','l','c','o','m','e','c')

print(tuple5.count('e'))
print(tuple5.index('c'))
```

```python
tuple6 = ('w','e','l','c','o','m','e')

#membership
print('c' in tuple6)
print('c' not in tuple6)
print('a' in tuple6)
print('a' not in tuple6)
```

```python
# iteration through tuple elements
tuple6 = ('w','e','l','c','o','m','e')
for letters in tuple6:
  print("letter is ->",letters)
```

```python
# built-in functions with tuple
tuple7 = (22,33,55,44,77,66,11)
print(tuple7)

print(max(tuple7))
print(min(tuple7))
print(sorted(tuple7))
print(len(tuple7))
```
