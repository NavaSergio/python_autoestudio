```python
# in
a = [1, 2, 3, 4]
print(4 in a)
print(44 in a)
print(44 not in a)
```

```python
# lambda
# permite definir funciones sin nombre
a = lambda x: x*2
for i in range(1,6):
  print(i,a(i))
```

```python
# nonlocal
def outer_function():
  a = 5
  def inner_function():
    nonlocal a
    a = 10
    print("Inner function:",a)
  inner_function()
  print("Outer function:",a)

outer_function()
```

```python
# return
def func_return():
  a = 10
  return a
print(func_return())
```

```python
# yield
def generator():
  for i in range(6):
    yield i*i

g = generator()
for i in g:
  print(i)
```
