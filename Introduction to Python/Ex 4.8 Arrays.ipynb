```python
# Defining and declaring and array
arr = [10, 20, 30, 40, 50]
print(arr)

# Accessing Array elements
print(arr[0])
print(arr[1])
print(arr[2])
print(arr[-1])
print(arr[-2])
```

```python
brands = ["Coke", "Apple", "Google", "Microsoft", "Toyota"]
print(brands)

# finding the length of the array
num_brands = len(brands)
print(num_brands)

# Adding an element to an array using append()
brands.append("Intel")
print(brands)
```

```python
colors = ["violet", "indigo", "blue", "green", "yellow", "orange", "red"]
print(colors)

# a continuación se elimina la entrada 4, recordar que comienza con el 0
del colors[4]
print(colors)

colors.remove("blue")
print(colors)

# a continuación se elimina la entrada 3, recordar que comienza con el 0
colors.pop(3)
print(colors)
```

```python
# Modifying elements of an array using indexing
fruits = ["Apple", "Banana", "Mango", "Grapes", "Orange"]

# a continuación modificaremos la entada 1 "Banana"  a "Pineapple"
fruits[1] = "Pineapple"
print(fruits)

# a continuación modificaremosla última entrada "Orange" a "Guava"
fruits[-1] = "Guava"
print(fruits)
```

```python
# Concatenando
concat = [1, 2, 3]
print(concat)

concat + [4, 5, 6]
print(concat)

concat = concat + [4, 5, 6]
print(concat)
```

```python
# Repitiendo elementos en un arreglo
repeat = ["a"]
print(repeat)

repeat = repeat * 5
print(repeat)
```

```python
# slicing an array
fruits = ["Apple", "Banana", "Mango", "Grapes", "Orange"]
print(fruits)

# nota importante desde:hasta, "desde" es "inclusivo" y "hasta" es "exclusivo"
print(fruits[1:4])
print(fruits[ :  3])
print(fruits[-4:])
print(fruits[-3:-1])
```

```python
# Declarando y definiendo un arreglo multidemensional
multd = [[1,2], [3,4], [5,6], [7,8]]
print(multd)
print(multd[0])
print(multd[3])
print(multd[2][1])
print(multd[3][0])
```
